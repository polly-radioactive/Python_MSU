## 1. Немного статистики

```python
# Запись чисел от 1000 до 10000 в файл
with open('input.txt', 'w') as file:
    numbers = [str(num) for num in range(1000, 10001)]  # Генерация чисел от 1000 до 10000
    file.write(' '.join(numbers))  # Запись чисел в файл, разделенных пробелом
```

```python
import numpy as np
with open('input.txt', 'r') as file:
    numbers = list(map(float, file.read().split()))

arr = np.array(numbers)

with open('output.txt', 'w') as file:
    file.write("{:.2f} {:.2f} {:.2f}".format(np.median(arr), np.mean(arr), np.std(arr)))
```

## 3. Две чугунные гири

```python
import pandas as pd
import numpy as np


num_days = 35

# нормальное распределение
mean1 = 30
std_dev1 = 4
data_type1 = np.random.normal(mean1, std_dev1, size=(10, num_days))
data_type1_int = np.round(data_type1).astype(int)

# случайное распределение
data_type2 = np.random.randint(20, 40, size=(20, num_days))
data_type2_int = np.round(data_type2).astype(int)

# # Создание DataFrame
df_type1 = pd.DataFrame(data_type1_int)
# df_type1['type'] = 1  # Добавление столбца для обозначения типа клиентов

df_type2 = pd.DataFrame(data_type2_int)
# df_type2['type'] = 2  # Добавление столбца для обозначения типа клиентов

df = pd.concat([df_type1, df_type2])
df.to_csv('input.csv', index=False)
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('input.csv')

count_1=0
count_2=0
for index, row in df.iterrows():
    dev=np.std(row)
    if dev <= 4:
      count_1 +=1
    else:
      count_2 +=1

if count_1 > count_2:
  print(1)
else:
  print(2)
```

## 4. Очистка логов

```python
import numpy as np
data = np.genfromtxt('input.csv', delimiter=',')
rows, cols = data.shape
data_new = data
for i in range(rows):
    for j in range(cols):
        if i % 2 == 0 and j % 2 != 0:
            data_new[i,j]=data[i, j] / 2
        if i % 2 != 0 and j % 2 == 0:
            data_new[i,j]=data[i, j] / 2

np.savetxt('output.csv', data_new, delimiter=',', fmt='%g')
```

## 5. Дуэльный клуб

```python
import numpy as np

data = np.genfromtxt('input.csv', delimiter=',')
daily_wins = np.sum(data, axis=0)
average_wins = daily_wins / 15

for i in range(30):
  data[0][i] = int(average_wins[i] * 1.5)

np.savetxt('output.csv', data, delimiter=',', fmt='%g')
```
