```python
def bin_search(lst):
    low = 0
    high = len(lst) - 1

    while low <= high:
        mid = (low + high) // 2
        if int(lst[mid]) == 1415:
            return mid
        elif int(lst[mid]) < 1415:
            low = mid + 1
        else:
            high = mid - 1
    return -1  # Если элемент не найден, возвращаем -1

lst=[1414, 1414, 1415]
print(bin_search(lst))
```

```python
import numpy as np

with open('/content/drive/MyDrive/Colab Notebooks/Python/lec_7/input.txt', 'r') as file:
    data = file.readlines()

start_profit = float(data[0])
end_profit = float(data[1])
days = int(data[2])
planned_profit = np.linspace(start_profit, end_profit, days)

for day in range(days):
    if day % 7 == 0:
        planned_profit[day] /= 3
    elif day % 7 == 4:
        planned_profit[day] *= 2


with open('/content/drive/MyDrive/Colab Notebooks/Python/lec_7/output.txt', 'w') as file:
    for profit in planned_profit:
        file.write(f"{profit:.2f}\n")
```
