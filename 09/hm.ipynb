## 1. Железная логика

```python
for x1 in False, True:
  for x2 in False, True:
    for x3 in False, True:
      if (not x1 and x2 and not x3) or (not x1 and x2 and x3) or (x1 and not x2 and not x3):
        result = f"{x1} {x2} {x3}"
        print(result)
```

## 2. Ошибка выжившего

```python
import csv
with open('aircraft_hits.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))

for i in range(len(data)):
    data[i] = [int(x) for x in data[i]]

averages = [sum(area) / len(data) for area in zip(*data)]
print(averages)
choice = averages.index(min(averages)) + 1
print(choice)
```

## 3. Электрик Станислав и интернет-казино

```python
import csv

with open('casino.csv') as csvfile:
    data = list(csv.reader(csvfile))
    numbers = [list(map(float, row)) for row in data[1:]]
    averages = [sum(area) for area in zip(*numbers)]
    print(averages)
    choice = averages.index(max(averages))
    print(data[0][choice])
```

```python
import pandas as pd
data = pd.read_csv('casino.csv')
print(data.columns[data.mean().argmax()])
```

## 4. Инвестор Геннадий и новый Мерседес
