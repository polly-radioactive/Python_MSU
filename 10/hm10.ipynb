## 1. Предобработка данных

```python
import pandas as pd
df = pd.read_csv('ex01.csv')
df = df.dropna(subset=['name'])
mean_score = df['score'].mean()
df['score'].fillna(mean_score, inplace=True)
df.to_csv('output.csv', index=True)
```

## 2. Температура

```python
import pandas as pd

df = pd.read_csv('ex02.csv')
df['temperature_f'] = df['temperature_c'].apply(lambda x: round(x * 9 / 5 + 32))
df.to_csv('output.csv', index=True)

```

## 4. Водители склада

```python
import pandas as pd
df = pd.read_csv('ex04.csv')
if not df.empty:
  df['Объем груза'] = df.apply(lambda row: row['Объем груза'] if row['Тип операции'] == 'Привоз' else -row['Объем груза'], axis=1)
  df = df.groupby('Фамилия водителя')['Объем груза'].sum().reset_index()
  df = df.sort_values(by=['Объем груза', 'Фамилия водителя'], ascending=[False, True])
  df.to_csv('output.csv', index=False, encoding="utf8")
else:
  with open("output.csv", 'w') as output_file:
    print("Фамилия водителя,Объем груза", file=output_file)

```

## 5. Лёгкий подъём

```python
import pandas as pd

df = pd.read_csv('ex05.csv')
if not df.empty:
  df['Маршрут'] = df['Город отправления'] + df['Город прибытия']
  df = df.groupby('Номер борта')['Маршрут'].nunique().rename('Уникальных маршрутов').reset_index()
  df = df.sort_values(['Уникальных маршрутов','Номер борта'], ascending=[False, True])
  df.to_csv('output.csv', encoding='utf-8', index=False)
else:
  with open("output.csv", 'w') as output_file:
    print("Номер борта,Уникальных маршрутов", file=output_file)
```
