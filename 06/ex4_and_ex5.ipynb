```python
def frequency_analysis(number):
    # Создаем словарь для хранения частоты встречаемости цифр
    frequency = {}

    # Проходимся по каждой цифре в числе
    for digit in number:
        # Если цифра уже есть в словаре, увеличиваем ее частоту на 1
        if digit in frequency:
            frequency[digit] += 1
        # Если цифры нет в словаре, добавляем ее со значением 1
        else:
            frequency[digit] = 1

    # Находим максимальную частоту
    max_frequency = max(frequency.values())

    # Находим цифры с максимальной частотой
    most_frequent_digits = [digit for digit, freq in frequency.items() if freq == max_frequency]

    return min(most_frequent_digits)  # Возвращаем наименьшую из самых часто встречающихся цифр

# Пример использования
number = input("Введите число: ")
most_frequent_digit = frequency_analysis(number)
print("Наиболее часто встречающаяся цифра:", most_frequent_digit)
```

```python
def unique_species():
    # Открываем файл input.txt
    with open('input_1.txt', 'r') as file:
        # Создаем пустой список для хранения видов животных
        species = []

        # Читаем построчно файл
        for line in file:
            # Разбиваем строку на отдельные значения
            values = line.split()

            # Получаем вид животного
            animal_species = values[1]

            # Если вид животного отсутствует в списке, добавляем его
            if animal_species not in species:
                species.append(animal_species)

        # Сортируем список видов животных по длине названия
        species.sort(key=lambda x: len(x))

        # Выводим виды животных через пробел
        print('\n'.join(species))

# Пример использования
unique_species()
```
