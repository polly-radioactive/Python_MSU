```python
from collections import defaultdict

# Читаем данные из файла
with open('input_3.txt', 'r') as file:
    animal_data = file.readlines()

# Создаем словарь для хранения количества животных каждого вида
animal_counts = defaultdict(int)

# Обрабатываем данные
for animal in animal_data:
    # Получаем вид животного
    species = animal.split()[1]
    # Увеличиваем счетчик для данного вида
    animal_counts[species] += 1

# Выводим результат в порядке убывания количества животных
for species, count in sorted(animal_counts.items(), key=lambda x: x[1], reverse=True):
    print(f"{species} - {count}")
```
