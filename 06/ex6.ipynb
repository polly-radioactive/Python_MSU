```python
# Читаем данные из файла
with open('input_2.txt', 'r') as file:
    animal_data = file.readlines()

# Создаем словарь для хранения информации о животных
animals = {}

# Обрабатываем данные
for animal in animal_data:
    animal_info = animal.split()
    animal_id = animal_info[0]
    species = animal_info[1]
    gender = animal_info[2]

    # Проверяем, является ли родившийся в зоопарке имеющийся в этом словаре
    if species in animals:
        animals[species].add(gender)

    # Если нет такого вида, создаем множество для пола и добавляем его
    else:
        animals[species] = {gender}

# Находим виды, где есть и мужской, и женский пол
reproducing_species = [species for species, genders in animals.items() if 'male' in genders and 'female' in genders]

# Выводим результат
if len(reproducing_species)==0:
  print(0)
else:
  for species in sorted(reproducing_species, key=len):
    print(species)
```
