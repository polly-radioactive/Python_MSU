```python
from collections import defaultdict

# Читаем данные из файла
with open('input_4.txt', 'r') as file:
    animal_data = file.readlines()

# Создаем словарь для хранения ID животных каждого вида
animal_ids = defaultdict(list)

# Обрабатываем данные
for animal in animal_data:
    # Получаем вид и ID животного
    animal_info = animal.split()
    species = animal_info[1]
    animal_id = animal_info[0]
    # Добавляем ID в список для данного вида
    animal_ids[species].append(animal_id)

# Выводим результат с сортировкой
for species, ids in sorted(animal_ids.items(), key=lambda x: (len(x[0]), x[0])):
    print(f"{species}: {', '.join(sorted(ids))}")
```
